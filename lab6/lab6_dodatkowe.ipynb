```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("materialy/fontanna1.jpg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("materialy/fontanna2.jpg", cv2.IMREAD_GRAYSCALE)

# tworzenie obiektu swift i wykrycie punktw i opisw
sift = cv2.SIFT_create()

kp1, desc1 = sift.detectAndCompute(img1, None)
kp2, desc2 = sift.detectAndCompute(img2, None)

# dopasowanie deskryptorw
bf = cv2.BFMatcher()
matches = bf.knnMatch(desc1, desc2, k=2)

# filtracja najlepszych dopasowa
best_matches = [[m] for m, n in matches if m.distance < 0.05 * n.distance]

result = cv2.drawMatchesKnn(
    img1, kp1, img2, kp2, best_matches, None, flags=2
)

plt.figure(figsize=(16, 8))
plt.imshow(result)
plt.axis('off')
plt.show()
```
